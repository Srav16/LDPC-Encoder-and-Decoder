```python
#to get part check matrix H
def mul_shift_right(x,k):
#x is input base matrix
#k elements in base_matrix -1 to len(base)-1 so -1 to shift 
  if k==-1:
    H= np.zeros(len(x))
  else:
    H = [x[k:] x[1:k]]
return H
```

```python
def mul_sh(x,k):
  # x: input block; k: -1 or shift; y: output

  end = len(x)
  y = []
  if k==-1:
    y = np.zeros_like(x) # an array of zeros with the same shape and type as a given array.
  elif k==0:
    y = x
  else:
    y[0:end-k] = x[k:end] 
    y[end-k+1:end+1] = x[0:k] #multiplication by shifted identity
  y = np.array(y)
  return y
```
